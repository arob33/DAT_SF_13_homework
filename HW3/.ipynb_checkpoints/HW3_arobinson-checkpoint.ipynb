```python
import pandas as pd
```

```python
bank = pd.read_csv('bank.csv', sep=';')
```

```python
bank.head()

#1. Data summary
# Some variables are strings (representing discrete data),
# others are numeric (representing discrete or continuous)
```

```python
#2. Encoding
# I chose to use pd.get_dummies rather than sklearn.OneHotEncoder because
# this is what we did in class and it is easier to implement given that
# we are already using a pandas dataframe

categorical_vars = ['job', 'marital', 'education', 'default', 'housing',
                    'loan', 'contact', 'poutcome']

# I excluded day, month because I thought this would be too much data 
# mining. When I leave them in, month_october is the second most
# important factor

categorical_dummies = [pd.get_dummies(bank[var], prefix=var)
    for var in categorical_vars]

# drop first dummy for each variable
categorical_dummies = [dummies.drop(dummies.columns[0], axis=1)
                      for dummies in categorical_dummies]
categorical_dummies = pd.concat(categorical_dummies, axis=1)
categorical_dummies.head()
```

```python
#3. Decision tree model

from sklearn.tree import DecisionTreeClassifier

numeric_vars = ['age', 'balance', 'campaign']
# excluding duration because this variable is not known prior to call

X = pd.concat([categorical_dummies,
               bank[numeric_vars]],
              axis=1).dropna(axis=0, how='any')
y = bank['y'].map({'yes':1, 'no':0})

treeclf = DecisionTreeClassifier(max_depth=3, random_state=1)
treeclf.fit(X, y)

pd.DataFrame({'feature':X.columns,
              'importance':treeclf.feature_importances_}).sort(
    'importance', ascending=False).head()
```

```python
#4. Cross validation

from sklearn.cross_validation import cross_val_score
treeclf = DecisionTreeClassifier(max_depth=3, random_state=1)
scores = cross_val_score(treeclf, X.values, y.values, cv=5)
print scores
scores.mean()
# mean cv score is 89%
```

```python
#5. Repeat with extended data set
bank2 = pd.read_csv('bank-additional-full.csv', sep=';')
bank2.dtypes
```

```python
categorical_vars = ['job', 'marital', 'education', 'default', 'housing',
                    'loan', 'contact', 'poutcome', 'day_of_week']

categorical_dummies = [pd.get_dummies(bank2[var], prefix=var)
    for var in categorical_vars]

# drop first dummy for each variable
categorical_dummies = [dummies.drop(dummies.columns[0], axis=1)
                      for dummies in categorical_dummies]
categorical_dummies = pd.concat(categorical_dummies, axis=1)


numeric_vars = ['age', 'campaign', 'pdays', 'previous',
                'emp.var.rate', 'cons.price.idx',
                'cons.conf.idx', 'euribor3m', 'nr.employed']
# excluding duration because this variable is not known prior to call

X = pd.concat([categorical_dummies,
               bank2[numeric_vars]],
              axis=1).dropna(axis=0, how='any')
y = bank2['y'].map({'yes':1, 'no':0})

treeclf = DecisionTreeClassifier(max_depth=3, random_state=1)
treeclf.fit(X, y)

pd.DataFrame({'feature':X.columns,
              'importance':treeclf.feature_importances_}).sort(
    'importance', ascending=False).head()
```

```python
treeclf = DecisionTreeClassifier(max_depth=3, random_state=1)
scores = cross_val_score(treeclf, X.values, y.values, cv=5)
print scores
scores.mean()

# mean score is 55%. The new model underperforms the old one.
# This looks like an example of data mining.
```
